```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('./data/processed_data.csv')
df
```

```python
df['date_day'] = pd.to_datetime(df['date_day'], format='%Y/%m/%d')  # 时间标准化
df.index = df['date_day']
del df['date_day']
df
```

```python
from pylab import rcParams
rcParams['font.sans-serif'] = ['SimHei'] 

choosed_time = '12:00:00'
data = df['2021-05-01':'2021-8-31'][choosed_time]
train = df['2021-05-01':'2021-08-20'][choosed_time]
test = df['2021-08-20':'2021-8-31'][choosed_time]
# train
%matplotlib inline
train.plot(figsize=(15,8), title= 'datetime', fontsize=18,label='训练集')  
test.plot(figsize=(15,8), title= 'datetime' ,fontsize=18,label='测试集')
plt.legend(shadow=True)
plt.xlabel('时间' ,fontsize=18)         #设置 x 轴文字标记
plt.ylabel('总有功功率（kw）' ,fontsize=18)
plt.title('所选择的数据集',fontsize=18)
plt.show()
```

```python
from statsmodels.tsa.api import ExponentialSmoothing  # 指数平滑
train = df['2021-05-01':'2021-08-20']
test = df['2021-08-20':'2021-8-31']

T_hat = test.copy()
fit1 = ExponentialSmoothing(np.asarray(train[choosed_time]), seasonal_periods=7, trend='mul', seasonal='mul').fit()
T_hat['Holt_Winter'] = fit1.forecast(len(test))

plt.figure(figsize=(16, 8))
plt.plot(train[choosed_time], label='测试集')
plt.plot(test[choosed_time], label='训练集')
plt.plot(T_hat['Holt_Winter'], label='Holt_Winters')
plt.legend(loc='best',fontsize=18,shadow=True)
plt.title('Holt_Winters预测')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
'''
    :y_true 真实值
    :y_pred 预测值
    均为数组的形式
'''
def mape(y_true, y_pred):
    return np.mean(np.abs((y_pred - y_true) / y_true)) * 100

rmse_holt_w = sqrt(mean_squared_error(test[choosed_time], T_hat['Holt_Winter']))
holt_w_mape = mape(test[choosed_time], T_hat['Holt_Winter'])
print("Holt-Winters的均方根误差为： ",rmse_holt_w)
print("Holt-Winters的平均绝对百分比误差为： ",holt_w_mape)
```

```python
#自相关图
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
D_train = train[choosed_time].diff().dropna()
# D_train.plot()
plot_acf(D_train).show()
plot_pacf(D_train).show()
```

```python
#自回归移动平均模型（ARIMA）
from statsmodels.tsa.arima.model import ARIMA

ts_ARIMA = train[choosed_time]
fit1_Arima = ARIMA(ts_ARIMA, order=(0, 1, 7)).fit()
Y_hat_ARIMA = fit1_Arima.predict(start="2021-08-20", end="2021-8-31", dynamic=True)

plt.figure(figsize=(16, 8))
plt.plot(train[choosed_time], label='测试集')
plt.plot(test[choosed_time], label='训练集')
plt.plot(Y_hat_ARIMA,label='ARIMA')
plt.legend(loc='best',fontsize=18,shadow=True)
plt.title('ARIMA预测',fontsize=18)
```

```python
rmse_arima = sqrt(mean_squared_error(test[choosed_time], Y_hat_ARIMA))
arima_mape = mape(test[choosed_time], Y_hat_ARIMA)
print("ARIMA的均方根误差为： ",rmse_arima)
print("ARIMA的平均绝对百分比误差为： ",arima_mape)
```

```python
fit1_Arima.summary()
```

```python
#SARIMA模型
import statsmodels.api as sm
y_sarima = test.copy()
fit1_sm = sm.tsa.statespace.SARIMAX(train[choosed_time], order=(2, 1, 4), seasonal_order=(0, 1, 1, 7)).fit()
y_sarima['SARIMA'] = fit1_sm.predict(start="2021-08-20", end="2021-8-31", dynamic=True)

plt.figure(figsize=(16, 8))
plt.plot(train[choosed_time], label='测试集')
plt.plot(test[choosed_time], label='训练集')
plt.plot(y_sarima['SARIMA'],label='SARIMA')
plt.legend(loc='best',fontsize=18,shadow=True)
plt.title('SARIMA预测',fontsize=18)
```

```python
rmse_sarima = sqrt(mean_squared_error(test[choosed_time], y_sarima['SARIMA']))
sarima_mape = mape(test[choosed_time], y_sarima['SARIMA'])
print("SARIMA的均方根误差为： ",rmse_sarima)
print("SARIMA的平均绝对百分比误差为： ",sarima_mape)
```
